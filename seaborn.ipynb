```python
df = pd.read_csv('dataset.csv', parse_dates=[3])
```

```python
df.head()
```

```python
df.rename(columns={'SystemCodeNumber': 'Location', 'LastUpdated': 'Timestamp'}, inplace=True)
df['Day'] = df.Timestamp.dt.date
df['Month'] = df.Timestamp.dt.month
df['Hour'] = df.Timestamp.dt.hour
```

```python
park = df[df.Location.isin(['Broad Street', 'NIA South'])]
```

```python
park.head()
```

```python
sns.set_style('dark')
```

```python
months =  [pd.datetime(2016, 10, 1),
           pd.datetime(2016, 11, 1),
           pd.datetime(2016, 12, 1)]
plt.rc('date.autoformatter', day='%b %Y');
```

```python
sns.lineplot(x=park.Day, y=park.Occupancy)
plt.xticks(months);
```

```python
sns.lineplot(x=park.Hour, y=park.Occupancy);
```

```python
sns.lineplot(x=park.Hour, y=park.Occupancy, n_boot=10, errorbar=('ci', 69));
```

```python
sns.lineplot(x=park.Hour, y=park.Occupancy, n_boot=10, errorbar=None);
```

```python
sns.lineplot(x=park.Hour, y=park.Occupancy, estimator='sum');
```

```python
sns.lineplot(x='Day', y='Occupancy', data=park, hue='Location',
             palette=['gray', 'red'],
             style='Location')
plt.xticks(months);
```

```python
sns.lineplot(x='Hour', y='Occupancy', data=park, estimator='sum', size='Location');
```
